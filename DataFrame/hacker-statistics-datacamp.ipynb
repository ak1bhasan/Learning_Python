üö∂ Question: Simulate a Random Walk

You will simulate a random walk using a for loop.

NumPy has already been imported as np.

Tasks:

Create a list called random_walk that starts with the value 0.

Write a for loop that runs 100 times.

Inside the loop:

Set step equal to the last element of random_walk (use index -1).

Roll a dice using np.random.randint(1, 7).

Update step using the rules:

If dice is 1 or 2 ‚Üí go one step down.

If dice is 3, 4, or 5 ‚Üí go one step up.

If dice is 6 ‚Üí roll the dice again and go up by that number of steps.

Append the updated step to random_walk.

Print the final random_walk list.

```python
# NumPy is imported, seed is set

# Initialize random_walk
random_walk = [0]

# Complete the ___
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[x]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = step - 1
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

ü§ï Question: Implement Clumsiness

Your random walk simulation is working well, but there‚Äôs one more twist:

You are a bit clumsy. There is a 0.5% chance that you fall down and return to step 0.

Tasks:

Modify the outer loop so that the random walk simulation runs 20 times instead of 5.

Inside the loop, generate a random float using np.random.rand().

If the random float is less than or equal to 0.005, reset step to 0.

Keep the rest of the random walk logic unchanged.

After running the simulation, observe how clumsiness affects the walks.

```python
# numpy and matplotlib imported, seed set

# clear the plot so it doesn't get cluttered if you run this many times
plt.clf()

# Simulate random walk 20 times
all_walks = []
for i in range(20) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= 0.005 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

üìä Question: Plot the Distribution of Endpoints

You‚Äôve simulated several random walks, but now it‚Äôs time to answer the big question:

What are the odds of reaching 60 steps?

To analyze this, you need to look at the final positions (endpoints) of all simulated walks.

Tasks:

Modify your simulation so that the random walk runs 500 times.

After converting all_walks to a NumPy array and transposing it (np_aw_t):

Select the last row of np_aw_t and store it in a variable called ends.

(This row contains the final step of each simulation.)

Use plt.hist() to create a histogram of ends.

Use plt.show() to display the histogram.

After plotting, observe the distribution and estimate how likely it is to reach 60 steps.

```python
# numpy and matplotlib imported, seed set

# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```

```python
num_success = np.sum(ends >= 60) / len(ends)
print(num_success)
```
