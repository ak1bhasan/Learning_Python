❓ Question

Row labels (indices) of a pandas DataFrame are automatically set to integers. You can customize them using a list of labels.

```python
import pandas as pd

# Build cars DataFrame
names = ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt']
dr =  [True, False, False, False, True, True, True]
cpc = [809, 731, 588, 18, 200, 70, 45]
cars_dict = { 'country':names, 'drives_right':dr, 'cars_per_cap':cpc }
cars = pd.DataFrame(cars_dict)
print(cars)

# Definition of row_labels
row_labels = ['US', 'AUS', 'JPN', 'IN', 'RU', 'MOR', 'EG']

# Specify row labels of cars
cars.index = row_labels

# Print cars again
print(cars)
```

```python
import pandas as pd

# Create the data
data = {
    "cars_per_cap": [809, 731, 588, 18, 200, 70, 45],
    "country": ['United States', 'Australia', 'Japan', 'India', 'Russia', 'Morocco', 'Egypt'],
    "drives_right": [True, False, False, False, True, True, True]
}

# Create row labels
row_labels = ['US', 'AUS', 'JPN', 'IN', 'RU', 'MOR', 'EG']

# Create DataFrame
cars = pd.DataFrame(data, index=row_labels)

# Save as CSV
cars.to_csv('cars.csv')

# Confirm by reading it back
cars_loaded = pd.read_csv('cars.csv', index_col=0)
print(cars_loaded)
```

❓ loc and iloc (1)

With loc and iloc, you can select specific rows and columns at the same time.

loc → label-based selection

iloc → integer position-based selection

The cars DataFrame has already been imported:

import pandas as pd
cars = pd.read_csv('cars.csv', index_col=0)


The DataFrame contains:

country

cars_per_cap

drives_right

Row labels:
US, AUS, JPN, IN, RU, MOR, EG

```python
# Import cars data
import pandas as pd
cars = pd.read_csv('cars.csv', index_col = 0)
print(cars)
# Print out observation for Japan
print(cars.loc["JPN"])

# Print out observations for Australia and Egypt
print(cars.loc[["AUS","EG"]])
```
