Scatter Plot (GDP vs Life Expectancy)
You are given two lists:

gdp_cap  GDP per capita of different countries (2007)

life_exp  life expectancy in those same countries (2007)

Your task:

Create a scatter plot of gdp_cap vs life_exp.

Set the x-axis to a logarithmic scale using plt.xscale('log').

Display the plot using plt.show().

```python
import matplotlib.pyplot as plt

# Example data
gdp_cap = [
    974.5803384, 5937.029526, 6223.367465,
    4797.231267, 12779.37964, 34435.36744,
    36126.49266, 29796.04834, 1391.253792, 3822.620945
]

life_exp = [
    43.828, 76.423, 72.301,
    42.731, 75.32, 81.235,
    79.829, 75.635, 64.062, 75.02
]

# Create scatter plot
plt.scatter(gdp_cap, life_exp)

# Put the x-axis on a logarithmic scale
plt.xscale('log')

# Show plot
plt.show()
```

```python

import matplotlib.pyplot as plt
year = [1950, 1970, 1990, 2010]
pop = [2.519, 3.692, 5.263, 6.972]


# Add more Data
year = [1800, 1850, 1900] + year
pop = [1.0, 1.262, 1.650] + pop

plt.plot(year, pop)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('World Population Projections')

plt.yticks([0, 2, 4, 6, 8, 10],
              ['0', '2B', '4B', '6B', '8B', '10B'])

plt.show()
```
